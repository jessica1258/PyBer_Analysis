### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
city_data_df.head()
```

```python
ride_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = fares_by_city_type/rides_by_city_type
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fares_per_driver = fares_by_city_type/drivers_by_city_type
avg_fares_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
#Creating dictionary of Series
pyber_sum={'Total Rides':pd.Series(rides_by_city_type),
       'Total Drivers': pd.Series(drivers_by_city_type),
       'Total Fares': pd.Series(fares_by_city_type),
       'Average Fare per Ride': pd.Series(avg_fare_per_ride),
       'Average Fare per Driver': pd.Series(avg_fares_per_driver),}
  
#Creating Dataframe
pyber_summary_df = pd.DataFrame(pyber_sum)
pyber_summary_df

```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
fpyber_summary_df=pyber_summary_df
fpyber_summary_df['Total Rides'] = fpyber_summary_df['Total Rides'].map('{:,.0f}'.format)
fpyber_summary_df['Total Drivers'] = fpyber_summary_df['Total Drivers'].map('{:,.0f}'.format)
fpyber_summary_df['Total Fares'] = fpyber_summary_df['Total Fares'].map('${:,.2f}'.format)
fpyber_summary_df['Average Fare per Ride'] = fpyber_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
fpyber_summary_df['Average Fare per Driver'] = fpyber_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)
fpyber_summary_df

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. (actually #0) Read the merged DataFrame 
pyber_data_df
```

```python
# 2. (actually #1) Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
data_group = pyber_data_df.groupby(["date","type"]).sum()["fare"]
data_group

```

```python
# 3. (actually #2) Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
data_group = data_group.reset_index()
data_group

```

```python
# 4. (actually #3) Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pyber_pivot_df = pd.DataFrame.pivot(data_group, index='date', columns='type', values='fare')

pyber_pivot_df.head(10)
```

```python
# 5. (actually #4) Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

pyber_select_pivot_df=pyber_pivot_df.loc["2019-01-01":"2019-04-29"]
pyber_select_pivot_df
```

```python
# 6. (actually #5) Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_select_pivot_df.index = pd.to_datetime(pyber_select_pivot_df.index)
pyber_select_pivot_df
```

```python
# 7. (actually #6) Check that the datatype for the index is datetime using df.info()
pyber_select_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
rs_pyber_df=pyber_select_pivot_df.resample('W').sum()
rs_pyber_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

ax=rs_pyber_df.plot(figsize=(20,6))


ax.set(xlabel=None)
ax.set(ylabel='Fare ($USD)')
ax.set_title('Total Fare by City Type',fontsize=20)

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
